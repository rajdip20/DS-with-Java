Given a linked list of size **n**, you have to **delete the node at position pos** of the linked list and return the new head. The position of initial node is 1.<br>
The tail of the circular linked list is connected to the head using next pointer.

<br>

**Example 1:**
>**Input:**<br>
>LinkedList: 1->2->3->4->5<br>
>(the first and last node are connected, i.e. 5 --> 1)<br>
>position: 4<br>
>**Output:** 1 2 3 5

**Example 2:**
>**Input:**<br>
>LinkedList: 2->5->7->8->99->100<br>
>(the first and last node are connected, i.e. 5 --> 1)<br>
>position: 6<br>
>**Output:** 2 5 7 8 99

<br>

**Your Task:**<br>
The task is to complete the function **deleteAtPosition()** which takes **head reference** and **pos** as argument and returns reference to the **new head** node, which is then used to display the list. The **printing** is done **automatically** by the **driver code**.

**Expected Time Complexity:** O(n).<br>
**Expected Auxiliary Space:** O(1).

**Constraints:**
- >2 <= number of nodes <= 10<sup>3</sup>
- >1 <= pos <= n

```python
'''
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
'''
def deleteAtPosition(head, pos):
    # code here
    def deleteHead(head):
        if not head:
            return None
        
        if head.next == head:
            return None
        
        last_node = head
        while last_node.next != head:
            last_node = last_node.next
        
        last_node.next = head.next
        next_head = head.next
        head.next = None
        head = next_head
        return head
    

    
    if not head:
        return None
    
    if head.next == head:
        return None
    
    if pos == 1:
        return deleteHead(head)
    
    prev = head
    for _ in range(pos - 2):
        prev = prev.next
    
    prev.next = prev.next.next
    
    return head
```
