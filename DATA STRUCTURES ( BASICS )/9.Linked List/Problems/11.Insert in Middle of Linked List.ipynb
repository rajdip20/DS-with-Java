Given a linked list of size **N** and a **key**. The task is to insert the key in the middle of the linked list.

<br>

**Example 1:**
>**Input:**<br>
>LinkedList = 1->2->4<br>
>key = 3<br>
>**Output:** 1 2 3 4<br>
>**Explanation:** The new element is inserted after the current middle element in the linked list.

**Example 2:**
>**Input:**<br>
>LinkedList = 10->20->40->50<br>
>key = 30<br>
>**Output:** 10 20 30 40 50<br>
>**Explanation:** The new element is inserted after the current middle element in the linked list and Hence, the output is 10 20 30 40 50.

<br>

**Your Task:**<br>
The task is to complete the function **insertInMiddle()** which takes head reference and element to be inserted as the arguments. The **printing** is done **automatically** by the **driver code**.

**Expected Time Complexity :** O(N)<br>
**Expected Auxilliary Space :** O(1)

**Constraints:**
- >1 <= N <= 10<sup>4</sup>

```python
'''
    Your task is to insert a new node in 
	the middle of the linked list with
	the given value.
	
	{
		# Node Class
		class Node:
		    def __init__(self, data):   # data -> value stored in node
		        self.data = data
		        self.next = None
	}
	
	Function Arguments: head (head of linked list), node 
	(node to be inserted in middle)
	Return Type: None, just insert the new node at mid.
'''
# Function to insert a node in the middle of the linked list.
def insertInMid(head, node):
    # code here
    if head == None:
        return node
    
    slow = head
    fast = head
    temp = head
    count = 0
    
    while temp:
        count += 1
        temp = temp.next
    
    while(fast and fast.next):
        prev = slow
        slow = slow.next
        fast = fast.next.next
    
    if count % 2 == 0:
        prev.next = node
        node.next = slow
    else:
        node.next = slow.next
        slow.next = node
```
